```python
import os
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split


    
```

```python
import csv
path = "./DischargingData/Discharging/Real/filter_result"
file_list = os.listdir(path)
file_list_csv = [file for file in file_list if file.endswith(".csv")]
file_length = len(file_list_csv)
ret = []
for file in file_list :
    data = pd.read_csv(path+'/'+file)
    ret.append(data)
df = pd.concat(ret)
print(df)
'''
ret = []
for k in range(file_length) :
    f = open(path+'/'+file_list_csv[k], 'r')
    wr = csv.reader(f)
    n = 0
    for line in wr :
        if n >= 1 : 
            ret.append(line)
        n = n+1
df = pd.DataFrame(ret, columns = ["Amp-hr","Amps","Voltage","Ro","Rp","Cp","Rdiff","Cdiff","SOC","Cycle","Temperature"])
#print(df)
'''

```

```python
#train = pd.read_csv(path+'/'+file_list_csv[k])
train = df
y = train.Temperature.values
train.drop(['Temperature'], inplace = True, axis = 1)
x = train.values

x, x_test , y, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify = y)

train_data = lgb.Dataset( x, label = y)
test_data = lgb.Dataset( x_test, label=y_test)

parameters = {
    #'application': 'binary',
    'objective': 'regression',
    'metric': 'mape',
    #'is_unbalance': 'true',
    'max_depth' : 10,
    'boosting': 'gbdt',
    #'num_leaves': 31,
    'feature_fraction': 0.7,
    #'bagging_fraction': 0.5,
    #'bagging_freq': 20,
    'learning_rate': 0.05,
    'num_iterations' : 1000,
    'verbose': 0
}
model = lgb.train(parameters,
                       train_data,
                       valid_sets=test_data,
                       num_boost_round=5000,
                       early_stopping_rounds=100)
#verbos_eval = 100
```

```python
from sklearn.metrics import mean_squared_error, r2_score
predict_train = model.predict(x)
predict_test = model.predict(x_test)
mse = mean_squared_error(y_test, predict_test)
r2 = r2_score(y_test, predict_test)
print('Mean squared error: ', mse)
print('R2 score: ', r2)
```
